**Reasoning**:
Now that the boxplots are displayed, I will calculate the first quartile (Q1), third quartile (Q3), and Interquartile Range (IQR) for 'Weekly_Sales', 'Temperature', 'Fuel_Price', and 'CPI', and then determine the lower and upper bounds for outlier detection using the IQR method, addressing instructions 3 and 4 of the subtask.



```python
columns_to_check = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI']
outlier_bounds = {}

for col in columns_to_check:
    Q1 = df_walmart[col].quantile(0.25)
    Q3 = df_walmart[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[col] = {'Q1': Q1, 'Q3': Q3, 'IQR': IQR, 'lower_bound': lower_bound, 'upper_bound': upper_bound}

    print(f"\n--- {col} ---")
    print(f"Q1: {Q1:.2f}")
    print(f"Q3: {Q3:.2f}")
    print(f"IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower_bound:.2f}")
    print(f"Upper Bound: {upper_bound:.2f}")

print("\nCalculated Q1, Q3, IQR, and outlier bounds for key columns.")
```

**Reasoning**:
Now that the outlier bounds are calculated, I will identify and count the number of outliers for each specified column ('Weekly_Sales', 'Temperature', 'Fuel_Price', and 'CPI') using these bounds, addressing instruction 5 and part of instruction 7 of the subtask.



```python
for col in columns_to_check:
    bounds = outlier_bounds[col]
    lower_bound = bounds['lower_bound']
    upper_bound = bounds['upper_bound']

    # Count outliers below the lower bound
    lower_outliers_count = df_walmart[df_walmart[col] < lower_bound].shape[0]
    # Count outliers above the upper bound
    upper_outliers_count = df_walmart[df_walmart[col] > upper_bound].shape[0]
    total_outliers_count = lower_outliers_count + upper_outliers_count

    print(f"\n--- Outlier Counts for {col} ---")
    print(f"Number of outliers below lower bound: {lower_outliers_count}")
    print(f"Number of outliers above upper bound: {upper_outliers_count}")
    print(f"Total outliers: {total_outliers_count}")

print("\nQuantitatively identified and counted outliers for key columns.")
```

**Reasoning**:
To investigate the correlation between extreme 'Weekly_Sales' spikes and holiday weeks, I will filter the `df_walmart` DataFrame to identify rows where 'Weekly_Sales' are considered extreme outliers (above the upper bound) and where 'IsHoliday' is True, as per instruction 6 of the subtask. Then, I will display the head and shape of these filtered holiday sales outliers.



```python
weekly_sales_upper_bound = outlier_bounds['Weekly_Sales']['upper_bound']
extreme_sales_outliers_on_holiday = df_walmart[(df_walmart['Weekly_Sales'] > weekly_sales_upper_bound) & (df_walmart['IsHoliday'] == True)]

print("Head of extreme Weekly_Sales outliers during holiday weeks:")
print(extreme_sales_outliers_on_holiday.head())
print("\n" + "-"*90)
print("Shape of extreme Weekly_Sales outliers during holiday weeks:", extreme_sales_outliers_on_holiday.shape)
print("\nIdentified extreme Weekly_Sales outliers that occurred on holiday weeks.")
```
